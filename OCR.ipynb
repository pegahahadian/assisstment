```python
# Detect Eq + Ans part
import cv2
import pytesseract
from pytesseract import Output
import matplotlib.pyplot as plt

img = cv2.imread('question.jpg')

# Convert the image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# pytesseract
custom_config = r'--oem 3 --psm 6'
details = pytesseract.image_to_data(gray, output_type=Output.DICT, config=custom_config)

# Iterate over each detected char
boxes = []
question_numbers = []
for i in range(len(details['text'])):
    if details['text'][i].startswith('#'):
        (x, y, w, h) = (details['left'][i], details['top'][i], details['width'][i], details['height'][i])
        boxes.append((x, y, w, h))
        question_numbers.append(details['text'][i].strip('#'))  # Extract question number

# Calculate the pixel dimensions for 2cm * 2.5cm
cm_to_pixels = 300 / 2.54 
width_2cm = int(5 * cm_to_pixels)
height_2point5cm = int(2 * cm_to_pixels)

# Adjust bounding boxes
for i in range(len(boxes)):
    x, y, w, h = boxes[i]
    center_x = x + w // 2 
    center_y = y + h // 2
    new_x = center_x - width_2cm // 2 
    new_y = center_y - height_2point5cm // 2 + int(0.3 * cm_to_pixels)
    boxes[i] = (int(new_x + 3 * cm_to_pixels), int(new_y), width_2cm, height_2point5cm)

# draw bounding boxes and labels around detected questions
def draw_boxes(img, boxes, question_numbers):
    for i, box in enumerate(boxes):
        x, y, w, h = box
        cv2.rectangle(img, (int(x), int(y)), (int(x + w), int(y + h)), (0, 255, 0), 2)
        cv2.putText(img, 'Q' + question_numbers[i], (int(x), int(y) - 10), cv2.FONT_HERSHEY_SIMPLEX, 1.1, (255, 0, 0), 2)
    return img

# Draw boxes and labels on the original image
boxed_img = draw_boxes(img.copy(), boxes, question_numbers)

# result image with boxes and labels
cv2.imwrite('detected_questions_with_boxes_and_labels.jpg', boxed_img)


plt.imshow(cv2.cvtColor(boxed_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
